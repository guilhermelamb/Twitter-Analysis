## Data Gathering

We will start from the easiest way to gather data and moving to the most challenging (at least in this work).

Let's first import all libraries that we will be using in this work

```python
import numpy as np
import pandas as pd
import requests
import os
import time
import json
import tweepy
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

The **first** data to be gathered is the .csv file twitter-archive-enhanced. This file contains lots of information on tweet posts of the We Rate Dogs page.

```python
twitter_archive = pd.read_csv('twitter-archive-enhanced.csv')
```

The **second** way that we will gather data is by programmatically downloading the file that contains the data about the dog's images.

```python
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'

r = requests.get(url)

with open(url.split('/')[-1], mode = 'wb') as file:
    file.write(r.content)
```

Now we can open the file 'image_predictions.tsv'

```python
image_predictions = pd.read_csv('image-predictions.tsv', sep = '\t')
```

The **third** way is by using the Twitter API and the library Tweepy. Here we will query the Twitter API for each tweet's JSON data, then we will store this information in a file, after that we will be able to open this file and transform it in a data frame.

```python
consumer_key = 'USER KEY'
consumer_secret = 'USER SECRET'
access_token = 'ACCESS TOKEN'
access_secret = 'ACCESS SECRET'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)

try:
    api.verify_credentials()
    print("Authentication OK")
except:
    print("Error during authentication")
    
    
start = time.time()
counter = 2356

with open('tweet_json.txt', 'w') as file:
    for tweet in twitter_archive['tweet_id']:
        try:
            data = api.get_status(tweet, tweet_mode='extended')
            json.dump(data._json, file)
            file.write('\n')
            counter -= 1
            print('{} loops left to finish'.format(counter))
            print('This is already running for {} seconds\n'.format(round(((time.time())-start),2)))
        except:
            counter -= 1
            print('Error! Tweet has been deleted!')
            print('{} loops left to finish'.format(counter))
            print('This is already running for {} seconds\n'.format(round(((time.time())-start),2)))
        
    

print('File saved!')
```

It took me almost 34 minutes to download this file! We will now open the tweet_json.txt file and read it line by line, after that we can transform it into a data frame.

```python
tweets_df = []

with open('tweet_json.txt', 'r') as file:
    for line in file:
        data = json.loads(line)
        tweets_df.append(data)
```

```python
tweet_json = pd.DataFrame(tweets_df)
```

## Data Cleaning

Before we start cleaning, let's create a copy of the data frames.

```python
twitter_archive_clean = twitter_archive.copy()
image_predictions_clean = image_predictions.copy()
tweet_json_clean = tweet_json.copy()
```

#### Define

***twitter_archive:***
- Change "tweet_id" to object;
- Drop rows that are replies and retweets;
- Drop columns with missing data;
- Change "timestamp" to datetime;
- Drop "source" column;
- Get the reference url for the tweets with missing url's;
- Get the right rating numerators;
- Drop rows with the denominator different from 10;
- Drop rows that do not have dog names;
- Split the dataframe into one for the tweets information and one for dogs information.

#### Coding & Testing

```python
#Change "tweet_id" to object
twitter_archive_clean['tweet_id']= twitter_archive_clean['tweet_id'].astype('str')
```

```python
#Change "timestamp to datetime
twitter_archive_clean['timestamp'] = pd.to_datetime(twitter_archive_clean['timestamp'])
```

```python
#Drop rows that are replies and retweets
drop1 = twitter_archive_clean[twitter_archive_clean['in_reply_to_status_id'].notnull()].index
twitter_archive_clean.drop(drop1, axis = 0, inplace = True)

drop2 = twitter_archive_clean[twitter_archive_clean['in_reply_to_user_id'].notnull()].index
twitter_archive_clean.drop(drop2, axis = 0, inplace = True)

drop3 = twitter_archive_clean[twitter_archive_clean['retweeted_status_id'].notnull()].index
twitter_archive_clean.drop(drop3, axis = 0, inplace = True)

drop4 = twitter_archive_clean[twitter_archive_clean['retweeted_status_user_id'].notnull()].index
twitter_archive_clean.drop(drop4, axis = 0, inplace = True)

drop5 = twitter_archive_clean[twitter_archive_clean['retweeted_status_timestamp'].notnull()].index
twitter_archive_clean.drop(drop5, axis = 0, inplace = True)
```

```python
#Drop columns with missing data ("in_reply_to_status_id", "in_reply_to_user_id", 
#"retweeted_status_id", "retweeted_status_user_id", "retweeted_status_timestamp")
twitter_archive_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id',
                              'retweeted_status_user_id', 'retweeted_status_timestamp'], axis = 1, inplace = True)
```

```python
#Drop "source" column, since it does not bring any usefull information in this study
twitter_archive_clean.drop('source', axis = 1, inplace = True)
```

```python
#Get the reference url for the tweets with missing url's
#Since the reference url follows a pattern "https://twitter.com/dog_rates/status/" + the tweet id, we can try to get this link
twitter_archive_clean[twitter_archive_clean['expanded_urls'].isnull()] 
```

After manually assessing the above tweet links (with the formula: https://twitter.com/dog_rates/status/ + tweet id), we can see that they all do not have photos, since this is not wanted in our project, we can drop this rows.

```python
twitter_archive_clean.dropna(axis = 0, subset = ['expanded_urls'], inplace = True)
```

```python
twitter_archive_clean[twitter_archive_clean['expanded_urls'].isnull()]
```

```python
#Drop rows with the denominator different from 10, some of these are right but most have have the wrong rating
#it's easier for us if we just keep the rating denominator at 10, since it's the vast majority.
twitter_archive_clean[twitter_archive_clean['rating_denominator'] != 10]
```

```python
#By looking at the dog names set, we can see that all non dog names are lower case and 'None', so we create the list bellow to
#drop all rows with these words.
#Manually assessing some tweets where name = None, I saw that the data frame has 673 rows where this condition is true,
#they are real tweets just missing the dog's name. Deleting this would have a huge impact on the data frame, so I excluded
#'None from the 'dog_names_to_delete' list.

dog_names_to_delete = ['a', 'actually', 'all', 'an', 'by', 'getting', 'his', 'incredibly', 'infuriating',
                       'just', 'life', 'light', 'mad', 'my', 'not', 'officially', 'old', 'one', 'quite', 'space',
                       'such', 'the', 'unacceptable', 'very']
```
